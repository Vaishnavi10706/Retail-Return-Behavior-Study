## Data Cleaning

```python
# filled description with No description
df['Description'] = df['Description'].fillna("No Description")
```

```python
df.isnull().sum()
```

## Feature Engineering

```python
# column TotalPrice
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
df
```

```python
df['InvoiceDate'].head()
```

```python
df = df.copy()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], dayfirst=True)
```

```python
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
```

```python
df['Return'] = df['Quantity'].apply(lambda x: 1 if x < 0 else 0)
```

```python
df
```

```python
# To check class imbalance
sns.countplot(data=df, x='Return')
plt.title('Return vs Not Return')
plt.show()
```

```python
# Compare: Returned vs Non-returned items
sns.boxplot(x='Return', y='Quantity', data=df)
plt.title('Quantity vs Return')
plt.show()
```

```python
# See if: High-price items → more returns
sns.boxplot(x='Return', y='UnitPrice', data=df)
plt.title('Unit Price vs Return')
plt.show()
```

```python
# Helps detect spending pattern of returns
sns.boxplot(x='Return', y='TotalPrice', data=df)
plt.title('Total Price vs Return')
plt.show()
```

```python
# Seasonal trends
sns.countplot(x='Month', hue='Return', data=df)
plt.title('Month vs Return')
plt.show()
```

```python
# Trend over years
sns.countplot(x='Year', hue='Return', data=df)
plt.title('Year vs Return')
plt.show()
```

```python
# Get top 10 countries
top10 = df['Country'].value_counts().head(10).index

# Plot only top 10
sns.countplot(
    y='Country',
    hue='Return',
    data=df[df['Country'].isin(top10)],
    order=top10
)

plt.title('Top 10 Countries vs Return')
plt.show()
```
